```python
# Replacing the qunatitative values with their mean and num_of_doors based on frequency, drop price column
# calculate mean
avg_normalized_loss = df['normalized-losses'].astype('float').mean()
print("Average of normalized losses is", avg_normalized_loss)
```

```python
df['bore'].replace(np.NaN,avg_bore,inplace = True)
```

```python
avg_stroke = df['stroke'].astype('float').mean()
print("Average Stroke is", avg_stroke)
```

```python
df['stroke'].replace(np.NaN,avg_stroke,inplace = True)
```

```python
avg_horsep = df['horsepower'].astype('float').mean()
print("Average Horse power is",avg_horsep)
```

```python
df['horsepower'].replace(np.NaN,avg_horsep,inplace = True)
```

```python
avg_peak= df['peak-rpm'].astype('float').mean()
print("Average peak-rpm is",avg_peak)
```

```python
df['peak-rpm'].replace(np.NaN,avg_peak,inplace = True)
```

```python
df[['bore','stroke','price','horsepower','peak-rpm']] = df[['bore','stroke','price','horsepower','peak-rpm']].astype('float')
```

```python
df['normalized-losses'] = df['normalized-losses'].astype('int')
```

```python
df.dtypes
```

```python
df['city-L/100km'] = 235/df['city-mpg']
```

```python
df
```

```python
df['highway-L/100km'] = 235/df['highway-mpg']
```

```python
df.head()
```

```python
df.drop(columns = ['city-mpg','highway-mpg'], inplace = True)
```

```python
df.head()
```

```python
# normalizing the columns length, width, height using feature scaling
df['length'] = df['length']/max(df['length'])
```

```python
df['width'] = df['width']/max(df['width'])
df['height'] = df['height']/max(df['height'])
```

```python
df[['length','width','height']].head()
```

```python
# converting horsepower into bins for easy analysis
df['horsepower'] = df['horsepower'].astype('int',copy = True)
```

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
```

```python
plt.pyplot.hist(df['horsepower'], bins = 3)
plt.pyplot.xlabel('Horsepower')
plt.pyplot.ylabel('Count of HP')
plt.pyplot.title('Horsepower bins')
```

```python
bins = np.linspace(min(df['horsepower']),max(df['horsepower']),num = 4)
print(bins)
```

```python
group = ['low','medium','High']
df['hp_category'] = pd.cut(df['horsepower'],bins,labels = group, include_lowest = True)
df[['horsepower','hp_category']].head()
```

```python
df['hp_category'].value_counts()
```

```python
plt.pyplot.hist(df['hp_category'], bins = 3)
plt.pyplot.xlabel('Horsepower Category')
plt.pyplot.ylabel('Count of HP')
plt.pyplot.title('Horsepower bins')
```
